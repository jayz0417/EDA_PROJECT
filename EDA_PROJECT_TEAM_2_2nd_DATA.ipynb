```python
# 모듈 호출
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
import re 
import time
import random
import datetime
```

```python
compile_name = re.compile("\"productName\"\:\".*?\"")
#정규표현식으로 제품 이름을 찾습니다 
compile_1 = re.compile("\"category1Name\"\:\".*?\"")
#정규표현식으로 첫번째 카테고리(소분류)를 찾습니다 
compile_2 = re.compile("\"category2Name\"\:\".*?\"")
#정규표현식으로 두번째 카테고리(소분류)를 찾습니다 
compile_3 = re.compile("\"category3Name\"\:\".*?\"")
#정규표현식으로 세번째 카테고리(소분류)를 찾습니다 
compile_4 = re.compile("\"category4Name\"\:\".*?\"")
#정규표현식으로 네번째 카테고리(소분류)를 찾습니다 
compile_review = re.compile("\"reviewCount\":\d*")
#정규표현식으로 리뷰 수를 찾습니다 
compile_pur = re.compile("\"purchaseCnt\"\:\".*?\"")
#정규표현식으로 구매 수를 찾습니다 
compile_price = re.compile("\"mobileLowPrice\":\d*")
#정규표현식으로 가격을 찾습니다 
compile_date = re.compile("\"openDate\"\:\".{8}")
#정규표현식으로 등록일을 찾습니다 
compile_brand = re.compile("\"brand\"\:\".*?\"")
#정규표현식으로 브랜드를 찾습니다 
compile_tag = re.compile("\"manuTag\"\:\".*?\"")
#정규표현식으로 제품 태그를 찾습니다 
```

```python
# 크롤링 페이지 설정
```

```python
def gen_url(page_num) : 
    base = "https://search.shopping.naver.com/search/all?catId=50000024&frm=NVSHCAT&origQuery=%EB%8B%A4%EC%9D%B4%EC%96%B4%ED%8A%B8%EC%8B%9D%ED%92%88"
    param_start = "&pagingIndex=" + str(page_num)
    back = "&pagingSize=40&productSet=total&query=다이어트식품&sort=review&timestamp=&viewType=list"
    print (str(n) +"페이지 크롤링 진행중 [%s]" % datetime.datetime.now())
    return base + param_start + back
```

```python
# 크롤링 - 데이터 수집 형태
```

```python
def get_fields(page_source) : 
    cate_1 = []
    cate_2 = []
    cate_3 = []
    cate_4 = []
    list_name = []
    list_review = []
    list_price = []
    list_date = []
    list_brand = []
    list_tag = []
    
    
    name_list = compile_name.findall(page_source)
    category_1 = compile_1.findall(page_source)
    category_2 = compile_2.findall(page_source)
    category_3 = compile_3.findall(page_source)
    category_4 = compile_4.findall(page_source)
    review_count = compile_review.findall(page_source)
    price = compile_price.findall(page_source)
    date = compile_date.findall(page_source)
    brand = compile_brand.findall(page_source)
    tag = compile_tag.findall(page_source)


    for i in range(len(name_list)) :
        list_name.append(name_list[i].split(":")[1].replace("\"", ""))

    for i in range(len(category_1)) :
        cate_1.append(category_1[i].split(":")[1].replace("\"", ""))
    
    for i in range(len(category_2)) :
        cate_2.append(category_2[i].split(":")[1].replace("\"", ""))
    
    for i in range(len(category_3)) :
        cate_3.append(category_3[i].split(":")[1].replace("\"", ""))
    
    for i in range(len(category_4)) :
        cate_4.append(category_4[i].split(":")[1].replace("\"", ""))
  
    for i in range(len(review_count)) :
        list_review.append(int(review_count[i].split(":")[1]))

    for i in range(len(price)) :
        list_price.append(int(price[i].split(":")[1].replace("\"", "")))

    for i in range(len(date)) :
        list_date.append(int(date[i].split(":")[1].replace("\"", "")))

    for i in range(len(brand)) :
        list_brand.append(brand[i].split(":")[1].replace("\"", ""))
    
    for i in range(len(tag)) :
        list_tag.append(tag[i].split(":")[1].replace("\"", ""))
  
    df = pd.DataFrame(
    {
        "name" : list_name, 
        "price" : list_price,
        "brand" : list_brand, 
        "cate_1" : cate_1,
        "cate_2" : cate_2,
        "cate_3" : cate_3,
        "cate_4" : cate_4,
        "review" : list_review,
        "date" : list_date,
        "tag" : list_tag
        })
    
    return df
```

```python
# 크롤링 진행
```

```python
medi_result = []

for n in range(1, 201) :
    url = gen_url(n)
    page_source = requests.get(url).text
    result = get_fields(page_source)
    
    medi_result.append(result) 
    time.sleep(random.uniform(1,3))
```

```python
# 크롤링 처리된 데이터 합치기
```
