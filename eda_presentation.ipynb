## 시장 개요 

- 2020년 국내 건강기능식품 시장은 다양한 브랜드의 시장 진입 및 코로나19 확산에 따른 건강 관심 증대에 힘입어 4조9000억원 규모를 형성
- 성장을 주도하는 것은 프로바이오틱스(유산균)이며, 눈 건강, 피부 건강 등 새 기능성 원료들이 포함된 기타 시장도 성장하고 있어, 
- 건강 기능식품 시장이 확장됨과 동시에 다양화되고 있음을 나타냄
- 인기 기능성 원료가 전체 시장의 64.5%를 차지함 (홍삼, 프로바이오틱스, 비타민(종합 및 단일 포함), 오메가 3
- 참고로, 오픈서베이의 조사에 따르면 온라인 구매채널 이용률은 네이버쇼핑이 34.8%로 가장 높음 

[출처] 2020 건강기능식품 시장현황 및 소비자 실태 조사, 오픈서베이

## (험난했던) 전처리

### 1. "기타" 소분류가 너무 많아서 새로운 카테고리 신설, 재지정
     - 등록된 제품명을 다 쪼개서 성분명을 발굴하고 
     - 제품명에 해당 성분명이 있을 경우 성분명으로 카테고리 지정 
     - 마이너한 성분들(소수만 발견되는 경우)은 제외 
     - ex. 간영양제, 밀크씨슬, milkthistle, 밀크시슬, 밀크 씨슬, 밀크 시슬 -> 밀크씨슬
### 2. 소분류가 잘못 지정된 경우 & 지정되지 않은 경우 카테고리 지정
     - ex. 프로바이오틱스인데 기타에 들어가 있다거나, 오메가3인데 카테고리가 없다거나... 
### 3. 건강기능식품/다이어트식품이 아닌 제품들 제외 
     - ex. 제티 초코, 쿨피스, 심지어 여성 의류나 가전제품이 이 카테고리에 들어와있는 경우도 있었음 
### 4. 여러 카테고리의 제품을 모아서 파는 경우, 대표 상품을 특정할 수 없어 제외
     - ex. 추석맞이 선물 모음전, 홈쇼핑 핫딜전 등 
### 5. 소분류가 없는 경우, 중분류로 소분류를 대체 
     - ex. 홍삼, 건강분말, 건강음료, 건강환/정, 꿀 등은 하위 소분류(네번째 카테고리) 없이 
     - 중분류가 마지막 분류로 끝나서, 
     - 세부 소분류가 없는게 맞는지 네이버 쇼핑에서 확인 후 하위 소분류 = 중분류로 설정
