##### title

- 네이버쇼핑에서 판매가 많은 제품과 그렇지 않은 제품은 왜 잘 판매되지 않을까?

- 가설1. 판매가 많은 제품은 x 때문이고, 판매가 적은 제품은 x 때문이다.
    - 판매가 많은 제품과 적은 제품의 변수간의 관계에 대한 분석
- 가설2. 판매가 적은 제품이 판매를 많이 하기 위해서는 x의 영향이 크다
    - 판매가 적은 제품이 판매가 많이 된 때의 x와 판매를 많이 하기 위한 x 간의 적합도 분석
- 사례1. 실제로 판매가 적었지만, 영향을 주는 x를 바꿔 판매가 많아진 제품과 판매가 적었던 시기와 많은 시기
- 사례2. 논문에서 보여주는 사례

##### 결론
- 네이버쇼핑에서 판매가 많은 제품은 ~한 특징이 있고 (그래프)
- 네이버쇼핑에서 판매가 적은 제품은 ~한 특징이 있다 (그래프)
- 판매가 적은 제품을 판매가 많은 제품으로 만들기 위해서는 x가 중요하다(그래프)
- list[판매가 적은 제품]들은 x를 바꾸면 판매를 높일 수 있다(사례, 입증, 그래프)
